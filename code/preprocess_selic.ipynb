```python
import pandas as pd
import numpy as np
```

```python
selic_1 = pd.read_excel("../data/selic_diaria_1.xlsx")
selic_2 = pd.read_excel("../data/selic_diaria_2.xlsx")

selic_1.columns = selic_1.iloc[0]
selic_2.columns = selic_2.iloc[0]

selic_2
```

```python
selic_1 = selic_1.drop(0)
selic_2 = selic_2.drop(0)
```

```python
selic_1.rename(columns={"Data":"Date",
                        "Taxa (% a.a.)":"Selic"},inplace=True)
selic_2.rename(columns={"Data":"Date",
                        "Taxa (% a.a.)":"Selic"},inplace=True)
```

```python
selic_1.set_index("Date",inplace=True)
selic_2.set_index("Date",inplace=True)
```

```python
# selic_2 = selic_2.loc[:,"Selic"]
# selic_1 = selic_1.loc[:,"Selic"]
selic_1
```

```python
selic = pd.concat([selic_1,selic_2])

selic.index = pd.to_datetime(selic.index)
selic.sort_index(inplace=True)

selic = pd.DataFrame(selic)

selic["Week"] = selic.index.week
selic["Weekday"] = selic.index.weekday

selic["CumProd"] = selic["Fator diário"].cumprod()

selic
```

```python
ibov = pd.read_csv("../data/weekly_ibov_return.csv",index_col="Date")

ibov.index = pd.to_datetime(ibov.index)
ibov["Weekday"] = ibov.index.weekday

ibov
```

```python
selic_weekly = selic[selic.Weekday==0]

teste = pd.concat([selic_weekly,ibov],axis=1)

selic_weekly = pd.DataFrame(teste["CumProd"])

selic_weekly["shift"] = selic_weekly.CumProd.shift(1)

selic_weekly["Selic"] = (selic_weekly["CumProd"]/selic_weekly["shift"])-1

selic_weekly.Selic.iloc[0] = selic_weekly.Selic.iloc[1]


selic = selic_weekly["Selic"]

selic
```

```python
selic.to_csv("../data/weekly_selic.csv")
```
