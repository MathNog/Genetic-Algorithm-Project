```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_all_weights_together(w,obj,pred,period):
    plt.figure(figsize=(15,8))
    title = "Plot of all weights together - %s with %s Predictor - %s"%(obj,pred,period)
    plt.title(title)
    plt.plot(w,label=w.columns)
    plt.legend()
    plt.show()
```

```python
def plot_sum_weights(w,obj,pred,period):
    
    df_sum = pd.DataFrame(index=w.index)
    for asset in w.columns:
        sum_weights = []
        for idx,row in w.iterrows():
            sum_weights.append(row.sum())
        sum_weights
        df_sum[asset] = sum_weights
    
    plt.figure(figsize=(15,8))
    title = "Plot of sum of all weights - %s with %s Predictor - %s"%(obj,pred,period)
    plt.title(title)
    plt.plot(df_sum,label=df_sum.columns)
    plt.legend()
    plt.show()
```

## Return and Risk with Naive Return Prediction

```python
w_train = pd.read_csv("../output/weights/weights_train_returnrisk_ret_naive.csv",index_col="Unnamed: 0")
w_test = pd.read_csv("../output/weights/weights_test_returnrisk_ret_naive.csv",index_col="Unnamed: 0")

w_train.index = pd.to_datetime(w_train.index)
w_test.index = pd.to_datetime(w_test.index)
```

```python
plot_all_weights_together(w_train,"Return & Risk","Naive","Train")
```

```python
plot_all_weights_together(w_test,"Return & Risk","Naive","Test")
```

```python
plot_sum_weights(w_train,"Return & Risk","Naive","Train")
```

```python
plot_sum_weights(w_test,"Return & Risk","Naive","Test")
```
