```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_all_weights_together(w,obj,pred,period):
    plt.figure(figsize=(15,8))
    title = "Plot of all weights together - %s with %s - %s"%(obj,pred,period)
    plt.title(title)
    plt.plot(w,label=w.columns)
    plt.legend()
    plt.show()
```

## Sharpe with Naive Return Prediction

```python
w_train = pd.read_csv("../output/weights/weights_train_sharpe_ret_naive.csv",index_col="Unnamed: 0")
w_test = pd.read_csv("../output/weights/weights_test_sharpe_ret_naive.csv",index_col="Unnamed: 0")

w_train.index = pd.to_datetime(w_train.index)
w_test.index = pd.to_datetime(w_test.index)
```

```python
plot_all_weights_together(w_train,"Sharpe","Naive","Train")
```

```python
plot_all_weights_together(w_test,"Sharpe","Naive","Test")
```

## Return and Risk with Naive Return Prediction

```python
w_train = pd.read_csv("../output/weights/weights_train_returnrisk_ret_naive.csv",index_col="Unnamed: 0")
w_test = pd.read_csv("../output/weights/weights_test_returnrisk_ret_naive.csv",index_col="Unnamed: 0")

w_train.index = pd.to_datetime(w_train.index)
w_test.index = pd.to_datetime(w_test.index)
```

```python
plot_all_weights_together(w_train,"Return & Risk","Naive","Train")
```

```python
plot_all_weights_together(w_test,"Return & Risk","Naive","Test")
```
