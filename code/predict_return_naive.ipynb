# Predict Return Naive

In this notebook, the predicted return for the next week is the last observed return

```python
import pandas as pd
```

```python
returns = pd.read_csv("../data/weekly_returns.csv",index_col="Date")
returns.index = pd.to_datetime(returns.index)

returns
```

```python
predicted_returns_naive = returns.shift(1)
predicted_returns_naive
```

```python
predicted_returns_naive.to_csv("../data/predicted_returns_naive.csv")
```
