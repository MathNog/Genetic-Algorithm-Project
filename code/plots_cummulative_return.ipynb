```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing benchmark  IBOV

```python
ibov_return = pd.read_csv("../data/weekly_ibov.csv",index_col="Date")
ibov_return.index = pd.to_datetime(ibov_return.index)
ibov_cummulative_return = ibov_return.cumsum()

train_ibov = ibov_cummulative_return[ibov_cummulative_return.index.year<2020]
test_ibov = ibov_cummulative_return[ibov_cummulative_return.index.year>=2020]
```

## Def plot functions

```python
def plot_cummulative_return_ibov(cum_ret,train,obj,pred):

    if train:
        title = "Comparative Return on Training Data - %s with %s predictor"%(obj,pred)
        ibov = train_ibov
    else:
        title = "Comparative Return on Test Data - %s with %s predictor"%(obj,pred)
        ibov = test_ibov

    plt.figure(figsize=(20,10))
    plt.plot(ibov,label="IBOV")
    plt.plot(cum_ret,label="Portfolio")
    plt.title(title)
    plt.legend()
    plt.show()
```

## Return & Risk with Naive Predictor

```python
cum_ret_train = pd.read_csv("../output/cummulative_returns/cummulative_returns_train_returnrisk_ret_naive.csv",index_col="Unnamed: 0")
cum_ret_test = pd.read_csv("../output/cummulative_returns/cummulative_returns_test_returnrisk_ret_naive.csv",index_col="Unnamed: 0")

cum_ret_train.index = pd.to_datetime(cum_ret_train.index)
cum_ret_test.index = pd.to_datetime(cum_ret_test.index)
```

```python
plot_cummulative_return_ibov(cum_ret=cum_ret_train,train=True,obj="Return & Risk",pred="Naive")
```

```python
plot_cummulative_return_ibov(cum_ret=cum_ret_test,train=False,obj="Return & Risk",pred="Naive")
```

## Sharpe with Naive Predictor

```python
cum_ret_train = pd.read_csv("../output/cummulative_returns/cummulative_returns_train_sharpe_ret_naive.csv",index_col="Unnamed: 0")
cum_ret_test = pd.read_csv("../output/cummulative_returns/cummulative_returns_test_sharpe_ret_naive.csv",index_col="Unnamed: 0")

cum_ret_train.index = pd.to_datetime(cum_ret_train.index)
cum_ret_test.index = pd.to_datetime(cum_ret_test.index)
```
